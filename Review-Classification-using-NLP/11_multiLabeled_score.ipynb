```python
label_list=[]
label_score=[]

for review in df['morphed_list']:
    clean_score = 0
    service_score = 0
    facility_score = 0
    safety_score = 0
    price_score = 0
    revisit_score = 0
    
    for item in review:
        for clean in clean_dict:
            if item==clean:
                clean_score += 1
        for service in service_dict:
            if item==service:
                service_score += 1
        for facility in facility_dict:
            if item==facility:
                facility_score += 1
        for safety in safety_dict:
            if item==safety:
                safety_score += 1
        for price in price_dict:
            if item==price:
                price_score += 1
        for revisit in revisit_dict:
            if item==revisit:
                revisit_score += 1
    
# 위생/청결:1, 서비스:2, 시설/편의:3, 안전:4, 가격만족도:5, 재방문/지인추천:6
        
    temp = [0, 0]
    if temp[1]<clean_score:
        temp = [1, clean_score]
    if temp[1]<service_score:
        temp = [2, service_score]    
    if temp[1]<facility_score:
        temp = [3, facility_score]
    if temp[1]<safety_score:
        temp = [4, safety_score]
    if temp[1]<price_score:
        temp = [5, price_score]
    if temp[1]<revisit_score:
        temp = [6, revisit_score]
    
    label_list.append(temp[0])
    label_score.append(temp[1])
df['labeled'] = label_list
df['label_score'] = label_score
```

```python
cnt = 0
for x in label_list:
    if x == [0,0]:
        cnt += 1
print(cnt)
```
