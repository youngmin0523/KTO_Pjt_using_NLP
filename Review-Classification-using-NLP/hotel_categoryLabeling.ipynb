```python
label_list=[]
label_score=[]

for review in df['형태소분석']:
    tour_score = 0
    food_score = 0
    business_score = 0
    nature_score = 0
    rest_score = 0
    exercise_score = 0
    
    for item in review:
        for tour in tour_dict:
            if item==tour:
                tour_score += 1
        for food in food_dict:
            if item==food:
                food_score += 1
        for business in business_dict:
            if item==business:
                business_score += 1
        for nature in nature_dict:
            if item==nature:
                nature_score += 1
        for rest in rest_dict:
            if item==rest:
                rest_score += 1
        for exercise in exercise_dict:
            if item==exercise:
                exercise_score += 1
            
    temp = [0, 0]
    if temp[1]<rest_score:
        temp = [1, rest_score]
    if temp[1]<tour_score:
        temp = [2, tour_score]
    if temp[1]<food_score:
        temp = [3, food_score]    
    if temp[1]<business_score:
        temp = [4, business_score]
    if temp[1]<nature_score:
        temp = [5, nature_score]
    if temp[1]<exercise_score:
        temp = [6, exercise_score]
    
    label_list.append(temp[0])
    label_score.append(temp[1])
df['labeled'] = label_list
df['label_score'] = label_score
```

```python
cnt = 0
for x in label_list:
    if x == [0,0]:
        cnt += 1
print(cnt)
```
