```python
import pandas as pd
pd.set_option('display.max_row', 1500) # 데이터프레임 크기 조절
pd.set_option('display.max_columns', 100)
import numpy as np
import os
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import random
import unicodedata
import urllib.request
import urllib
import sys
import re
import math

def scroll_down(driver):
  driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
  time.sleep(1)
```

## 예약자리뷰

```python
name = '비치스토리호텔 [한국관광 품질인증/Korea Quality]'

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

cnt = soup.find('div', 'biz_name_area').find('a').get_text()
cnt = cnt.split(' ')[2]
cnt = int(cnt)
print(cnt)

driver.find_element_by_xpath('//*[@id="content"]/div[2]/div[2]/div[1]/div[2]/a[1]').click()

html1 = driver.page_source
soup1 = BeautifulSoup(html1, 'html.parser')

while True:
    try:
        scroll_down(driver)
        time.sleep(1)
        driver.find_element_by_xpath('//*[@id="content"]/div/a').click()
        time.sleep(1)
    except:
        break

html1 = driver.page_source
soup1 = BeautifulSoup(html1, 'html.parser')

review_li = soup1.find_all('div', 'review_txt')
rate_li = soup1.find_all('span', 'score')

i = 0
for li in review_li:
    review = li.get_text().replace('\n', '')
    review2.append(review)
    name2.append(name)
    i += 1
print(i)

j = 0
k = 0
for li in rate_li:
    if j > 4:
        rate = li.get_text()
        rate2.append(rate)
        k += 1
    j += 1
print(k)
```
